```python
import os

def list_files(directory):
    """
    Return a list of all files in the given directory.
    """
    try:
        files = [f for f in os.listdir(directory) if os.path.isfile(os.path.join(directory, f))]
        return files
    except OSError as e:
        print(f"Error reading directory {directory}: {e}")
        return None

# Specify the directory path
directory_path = r"C:\Users\LEGACY\Desktop\SOON-pEMP\Data"

# Get the list of files in the directory
file_list = list_files(directory_path)

if file_list is not None:
    print(f"Files in {directory_path}:")
    for file in file_list:
        print(file)
```

```python
import pandas as pd
# Specify the directory containing CSV files
directory = r"C:\Users\LEGACY\Desktop\SOON-pEMP\Data"

# Initialize UID counter
uid_counter = 1

# Iterate through each file in the directory
for filename in os.listdir(directory):
    if filename.endswith(".csv"):
        file_path = os.path.join(directory, filename)

        # Read the CSV file
        df = pd.read_csv(file_path)

        # Drop the "Timestamp" column if it exists
        if "Timestamp" in df.columns:
            df = df.drop(columns=["Timestamp"])

        # Add a new "UID" column with continuous values
        df["UID"] = range(uid_counter, uid_counter + len(df))
        uid_counter += len(df)

        # Save the modified DataFrame back to the CSV file
        df.to_csv(file_path, index=False)

        print(f"Processed {filename}")

print("Processing completed.")
```

```python
import os
import pandas as pd

# Directory containing the CSV files
directory = r"C:\Users\LEGACY\Desktop\SOON-pEMP\Data"

# List to store individual DataFrames
dataframes = []

# Iterate through each file in the directory
for filename in os.listdir(directory):
    if filename.endswith(".csv"):
        file_path = os.path.join(directory, filename)
        df = pd.read_csv(file_path)
        dataframes.append(df)

# Concatenate all DataFrames into one
combined_data = pd.concat(dataframes, ignore_index=True)

# Save the combined data to a new CSV file
combined_data.to_csv(os.path.join(directory, "combined_data.csv"), index=False)

print("Combined data saved to 'combined_data.csv'")
```
